### Facebook Ad click prediction
You have been hired as a consultant to a start-up that is running a targetted marketing ads on facebook. The company wants to anaylze customer behaviour by predicting which customer clicks on the advertisement. Customer data is as follows: 

Inputs: 
- Name 
- e-mail 
- Country 
- Time on Facebook 
- Estimated Salary (derived from other parameters)

Outputs:
- Click (1: customer clicked on Ad, 0: Customer did not click on the Ad)

# Step 0: Libraries Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

# Step 1:  Import Dataset

```python
training_set = pd.read_csv('Facebook_Ads_2.csv', encoding='ISO-8859-1')
```

```python
# Show the data head!
training_set.head()
```

```python
training_set.info()
```

```python
training_set.describe()
```

```python
Y_LABEL='Clicked'
```

# STEP #2: EXPLORE/VISUALIZE DATASET


```python
def analyze_data(training_set, Y_LABEL):
    clicked = training_set[training_set[Y_LABEL]==1]
    not_clicked = training_set[training_set[Y_LABEL]==0]
    # Count the survived and deceased 
    print("Total =", len(training_set))

    print("Number of people who clicked =", len(clicked))
    print("Percentage clicked =", 1.*len(clicked)/len(training_set)*100.0, "%")

    print("Did not click =", len(not_clicked))
    print("Percentage who did not click =", 1.*len(not_clicked)/len(training_set)*100.0, "%")
    
def plot_count_plot(X_LABEL, Y_LABEL):
    # Bar Chart to indicate the number of people survived based on their class
    # If you are a first class, you have a higher chance of survival
    plt.figure(figsize=[6,12])
    plt.subplot(211)
    sns.countplot(x = X_LABEL, data = training_set)
    plt.subplot(212)
    sns.countplot(x = X_LABEL, hue = Y_LABEL, data=training_set)
    
def plot_scatterplot(X_LABEL1, X_LABEL2, Y_LABEL):
    sns.scatterplot(training_set[X_LABEL1], training_set[X_LABEL2], hue = training_set[Y_LABEL])
```

```python
analyze_data(training_set, Y_LABEL)
```

# STEP #3: PREPARE THE DATA FOR TRAINING/ DATA CLEANING 

```python
#Let's drop the emails, country and names (we can make use of the country later!)
training_set.drop(['Names', 'emails', 'Country'],axis=1,inplace=True)
```

```python
training_set.head()
```

```python
Y = training_set[Y_LABEL]
X= training_set.drop(Y_LABEL, axis=1)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

# STEP#4: MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
Y_train = Y_train.values
Y_test = Y_test.values
```

```python
Y_train
```

```python
X_test
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, Y_train)
```

# STEP#5: MODEL TESTING

```python
def print_regression_metrics(Y_test, Y_predict_test):
    print(f"Precision ={precision_score(Y_test, Y_predict_test)}")
    print(f"Recall ={recall_score(Y_test, Y_predict_test)}")
    print(f"F1 score ={f1_score(Y_test, Y_predict_test)}")
    print(classification_report(Y_test, Y_predict_test))
```

```python
Y_predict_train = classifier.predict(X_train)
Y_predict_train
```

```python
print_regression_metrics(Y_train, Y_predict_train)
```

```python
cm = confusion_matrix(Y_train, Y_predict_train)
sns.heatmap(cm, annot=True, fmt="d")
```

```python
Y_predict_test = classifier.predict(X_test) 
```

```python
Y_predict_test
```

```python
cm = confusion_matrix(Y_test, Y_predict_test)
sns.heatmap(cm, annot=True, fmt="d")
```

```python
print_regression_metrics(Y_test, Y_predict_test)
```

# STEP #6: VISUALIZING TRAINING AND TESTING DATASETS

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, Y_train

# Create a meshgrid ranging from the minimum to maximum value for both features

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
```

```python
# plot the boundary using the trained classifier
# Run the classifier to predict the outcome on all pixels with resolution of 0.01
# Colouring the pixels with 0 or 1
# If classified as 0 it will be magenta, and if it is classified as 1 it will be shown in blue 
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('magenta', 'blue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
```

```python
# plot all the actual training points
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('magenta', 'blue'))(i), label = j)
    
plt.title('Facebook Ad: Customer Click Prediction (Training set)')
plt.xlabel('Time Spent on Site')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
def plot_contour(X, Y, set_type="Training"):
    X_set, y_set = X_train, Y_train
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                         np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
    plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                 alpha = 0.75, cmap = ListedColormap(('magenta', 'blue')))
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                    c = ListedColormap(('magenta', 'blue'))(i), label = j)
    plt.title(f'Facebook Ad: Customer Click Prediction ({set_type} set)')
    plt.xlabel('Time Spent on Site')
    plt.ylabel('Estimated Salary')
    plt.legend()
    plt.show()

plot_contour(X_train, Y_train)
```

```python
plot_contour(X_test, Y_test, "Test")
```
