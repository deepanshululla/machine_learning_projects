# Problem statement

The sinking of the Titanic on April 15th, 1912 is one of the most tragic tragedies in history. The Titanic sank after colliding with an iceberg, killing 1502 out of 2224 passengers. The numbers of survivors were low due to the lack of lifeboats for all passengers and crew. Some passengers were more likely to survive than others, such as women, children, and upper-class. This case study analyzes what sorts of people were likely to survive this tragedy. The dataset includes the following: 

- Pclass:	Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd)
- Sex:    Sex	
- Age:    Age in years	
- Sibsp:	# of siblings / spouses aboard the Titanic	
- Parch:	# of parents / children aboard the Titanic	
- Ticket:	Ticket number	
- Fare:	Passenger fare	
- Cabin:	Cabin number	
- Embarked:	Port of Embarkation	C = Cherbourg, Q = Queenstown, S = Southampton


- Target class: Survived: Survival	(0 = No, 1 = Yes)

```python
def analyze_survival(training_set):
    survived = training_set[training_set[Y_LABEL]==1]
    not_survived = training_set[training_set[Y_LABEL]==0]
    # Count the survived and deceased 
    print("Total =", len(training_set))

    print("Number of Survived passengers =", len(survived))
    print("Percentage Survived =", 1.*len(survived)/len(training_set)*100.0, "%")

    print("Did not Survive =", len(not_survived))
    print("Percentage who did not survive =", 1.*len(not_survived)/len(training_set)*100.0, "%")
    
def plot_count_plot(X_LABEL, Y_LABEL):
    # Bar Chart to indicate the number of people survived based on their class
    # If you are a first class, you have a higher chance of survival
    plt.figure(figsize=[6,12])
    plt.subplot(211)
    sns.countplot(x = X_LABEL, data = training_set)
    plt.subplot(212)
    sns.countplot(x = X_LABEL, hue = Y_LABEL, data=training_set)

def plot_scatterplot(X_LABEL1, X_LABEL2, Y_LABEL):
    sns.scatterplot(training_set[X_LABEL1], training_set[X_LABEL2], hue = training_set[Y_LABEL])
```

# STEP #3: PREPARE THE DATA FOR TRAINING/ DATA CLEANING 

```python
# Let's explore which dataset is missing
sns.heatmap(training_set.isnull(), yticklabels = False, cbar = False, cmap="Blues")
```

```python
# drop unneccessary columns
training_set.drop(['Name', 'Ticket', 'Embarked', 'PassengerId', 'Cabin'],axis=1,inplace=True)
```

```python
training_set.head()
```

```python
X_train = training_set.copy().drop(Y_LABEL, axis=1)
```

```python
X_train.info()
```

```python
Y_train = training_set[Y_LABEL]
```

```python
# Let's view the data one more time!
sns.heatmap(training_set.isnull(), yticklabels = False, cbar = False, cmap="Blues")
```

```python
# Let's get the median age for male (~29) and female (~25)
plt.figure(figsize=(15, 10))
sns.boxplot(x=Y_LABEL, y='Age',data=training_set)
```

```python
def Fill_Age(data):
    age = data[0]
    sex = data[1]

    if pd.isnull(age):
        if sex=='male': 
            return 29
        else:
            return 25
    else:
        return age
```

```python
# fill nulls with median
training_set['Age'] = training_set[['Age','Sex']].apply(Fill_Age,axis=1)
```

```python
# Let's view the data one more time!
sns.heatmap(training_set.isnull(), yticklabels = False, cbar = False, cmap="Blues")
```

```python
# You just need one column only to represent male or female
pd.get_dummies(training_set['Sex'])
```

```python
is_male = pd.get_dummies(training_set['Sex'], drop_first = True)
```

```python
is_male.head()
```

```python
training_set.head()
```

```python
# Now let's add the encoded column male again
training_set = pd.concat([training_set, is_male], axis=1)
```

```python
# first let's drop the embarked and sex 
training_set.drop(['Sex'], axis=1, inplace=True)
```

```python
training_set.info()
```

```python
training_set.head()
```

# STEP#4: MODEL TRAINING

```python
#Let's drop the target coloumn before we do train test split
X = training_set.drop(Y_LABEL,axis=1)
Y = training_set[Y_LABEL]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
X_train.info()
```

```python
Y_train.info()
```

```python
X_train['male'].count()
```

```python
# Fitting Logistic Regression to the Training set

classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, Y_train)
```

# STEP#5: MODEL TESTING

```python
Y_predict_train = classifier.predict(X_train)
```

```python
cm = confusion_matrix(Y_train, Y_predict_train)
sns.heatmap(cm, annot=True, fmt="d")
```

```python
Y_predict_test = classifier.predict(X_test)
Y_predict_test
```

```python
cm = confusion_matrix(Y_test, Y_predict_test)
sns.heatmap(cm, annot=True, fmt="d")
```

```python
tn, fp, fn, tp = cm.ravel()
```

```python
print(tn, fp, fn, tp)
```

```python
precision_score(Y_test, Y_predict_test)
```

```python
# extra code  this cell also computes the precision: TP / (FP + TP)
cm[1, 1] / (cm[0, 1] + cm[1, 1])
```

```python
recall_score(Y_test, Y_predict_test)  
```

```python
# extra code  this cell also computes the recall: TP / (FN + TP)
cm[1, 1] / (cm[1, 0] + cm[1, 1])
```

```python
f1_score(Y_test, Y_predict_test)  
```

```python
# extra code  this cell also computes the f1 score
cm[1, 1] / (cm[1, 1] + (cm[1, 0] + cm[0, 1]) / 2)
```

```python
def print_regression_metrics(Y_test, Y_predict_test):
    print(f"Precision ={precision_score(Y_test, Y_predict_test)}")
    print(f"Recall ={recall_score(Y_test, Y_predict_test)}")
    print(f"F1 score ={f1_score(Y_test, Y_predict_test)}")
    print(classification_report(Y_test, Y_predict_test))
    
print_regression_metrics(Y_test, Y_predict_test)
```

```python
print_regression_metrics(Y_train, Y_predict_train)
```
